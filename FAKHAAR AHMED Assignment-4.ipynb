1. Create a class 'BankAccount' with attributes account holder name, account number, and
balance. Add methods for deposit and withdrawal. Implement logic to prevent withdrawal if balance
is insufficient.

```python
class BankAccount:
    def __init__(self, name, account_number, balance):
        self.name = name
        self.account_nunmber = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance -= amount
```

```python
acc = BankAccount("Ali", 123, 1000)
acc.deposit(500)
acc.withdraw(200)
print(acc.balance)
```

2. Design a class 'Student' with attributes name and marks. Create a method to calculate grade
based on marks using conditions (A, B, C, Fail).

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def grade(self):
        if self.marks >= 80:
            return "A"
        elif self.marks >= 60:
            return "B"
        elif self.marks >= 40:
            return "C"
        else:
            return "Fail"
```

```python
s = Student("Ali", 75)
print(s.grade())
```

3. Create a class 'ElectricityBill' for a KE-electric consumer. Include units consumed and calculate
bill based on slab rates (e.g., different price per unit for different ranges).

```python
class ElectricityBill:
    def __init__(self, units):
        self.units = units

    def calculate_bill(self):
        if self.units <= 100:
            bill = self.units * 10
        elif self.units <= 200:
            bill = self.units * 15
        else:
            bill = self.units * 20
        return bill
```

```python
e = ElectricityBill(150)
print("Bill =", e.calculate_bill())
```

4. Design a class 'Vehicle' and create subclasses 'Car' and 'Bike'. Override a method
'start_engine()' differently in each subclass to demonstrate polymorphism.

```python
class Vehicle:
    def start_engine(self):
        print("Starting vehicle engine...")
```

```python
class Car(Vehicle):
    def start_engine(self):
        print("Car engine started with a key or push button.")
class Bike(Vehicle):
    def start_engine(self):
        print("Bike engine started with a self-start or kick-start.")
vehicles = [Car(), Bike()]

for vehicle in vehicles:
    vehicle.start_engine()
```

5. Create a class 'Employee' with basic salary. Add a method to calculate total salary including
bonus (10%) and tax deduction (5%).

```python
class Employee:
    def __init__(self, salary):
        self.salary = salary

    def total_salary(self):
        return self.salary + (self.salary * 0.10) - (self.salary * 0.05)
```

```python
e = Employee(500)
print(e.total_salary())
```

6. Design an abstract class 'Shape' with a method 'area()'. Create subclasses 'Circle' and
'Rectangle' and implement area calculation logic.

```python
class Shape:
    def area(self):
        pass
```

```python
class Circle(Shape):
    def __init__(self, r):
        self.r = r

    def area(self):
        return 3.14 * self.r * self.r
```

```python
class Rectangle(Shape):
    def __init__(self, l, w):
        self.l = l
        self.w = w

    def area(self):
        return self.l * self.w
```

```python
c = Circle(5)
r = Rectangle(4, 6)

print(c.area())
print(r.area())
```

7. Create a class 'HospitalPatient' with attributes name, age, and disease. Add a method to
determine priority (Emergency if age > 60 or serious disease).

```python
class HospitalPatient:
    def __init__(self, name, age, disease):
        self.name = name
        self.age = age
        self.disease = disease

    def determine_priority(self):
        serious_diseases = {"heart attack", "stroke", "cancer", "respiratory failure"}

        if self.age > 60 or self.disease.lower() in serious_diseases:
            return "Emergency"
        else:
            return "Normal"
```

```python
p1 = HospitalPatient("Ali", 70, "flu")
p2 = HospitalPatient("Sara", 30, "heart attack")

print(p1.determine_priority()) 
print(p2.determine_priority())  
```

8. Build a class 'ShoppingCart' where user can add items with price. Implement logic to calculate
total bill and apply discount if total exceeds a certain amount.

```python
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, name, price):
        self.items.append(price)

    def total_bill(self):
        total = sum(self.items)

        if total > 1000:   # discount condition
            total = total * 0.9  # 10% discount

        return total
```

```python
cart = ShoppingCart()
cart.add_item("Shoes", 800)
cart.add_item("Shirt", 300)

print(cart.total_bill())
```

9. Create a class 'MobilePhone' with private attribute battery percentage. Use getter and setter
methods to update battery with validation (0–100 only).

```python
class MobilePhone:
    def __init__(self, battery):
        self.__battery = 0
        self.set_battery(battery)

    def get_battery(self):
        return self.__battery

    def set_battery(self, value):
        if 0 <= value <= 100:
            self.__battery = value
        else:
            print("Invalid battery value! Must be between 0 and 100.")
```

```python
phone = MobilePhone(50)

print(phone.get_battery())  # 50

phone.set_battery(120)      # Invalid battery value!
print(phone.get_battery())  # still 50
```

10. Design a class 'Library' where users can borrow and return books. Add logic to check
availability before issuing a book.

```python
class Library:
    def __init__(self):
        self.books = {}

    def add_book(self, name, qty):
        self.books[name] = self.books.get(name, 0) + qty

    def borrow_book(self, name):
        if self.books.get(name, 0) > 0:
            self.books[name] -= 1
            print("Borrowed", name)
        else:
            print("Not available")

    def return_book(self, name):
        self.books[name] = self.books.get(name, 0) + 1
        print("Returned", name)
```

```python
lib = Library()
lib.add_book("Python Basics", 3)
lib.borrow_book("Python Basics")
lib.return_book("Python Basics")
```
