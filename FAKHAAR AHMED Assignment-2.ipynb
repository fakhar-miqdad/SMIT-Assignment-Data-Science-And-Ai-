# User se marks lo aur grade do:

# 90+ → A
# 80+ → B
# 70+ → C
# 60+ → D
# warna → Fail

```python
marks = int(input("Enter the Marks : "))

if marks >= 90:
    print("Grade: A")
elif marks >= 80:
    print("Grade: B")
elif marks >= 70:
    print("Grade: C")
elif marks >= 60:
    print("Grade: D")
else:
    print("Fail")
```

# User se day number (1–7) lo:

# 1 → Monday
# 2 → Tuesday
## ...
# 7 → Sunday

```python
day = int(input("Enter the day number (1-7) : "))

if day == 1:
    print("Monday")
elif day == 2:
    print("Tuesday")
elif day == 3:
    print("Wednesday")
elif day == 4:
    print("Thursday")
elif day == 5:
    print("Friday")
elif day == 6:
    print("Saturday")
elif day == 7:
    print("Sunday")
else:
    print("Invalid input")
```

# User se month number lo:

# 12,1,2 → Winter
# 3,4,5 → Spring
# 6,7,8 → Summer
# warna → Autumn

```python
month = int(input("Enter the Month Number (1-12) : "))

if month in [12, 1, 2]:
    print("Winter")
elif month in [3, 4, 5]:
    print("Spring")
elif month in [6, 7, 8]:
    print("Summer")
elif month in [9, 10, 11]:
    print("Autumn")
else:
    print("Invalid input")
```

# Simple calculator:
# User se operator lo (+, -, *, /):

# + → addition
# _ → subtraction
# * → multiplication
# / → division
# warna → "Invalid operator"

```python
a = int(input("Enter 1st number : "))
b = int(input("Enter 2nd number : "))
op = input("Enter the Operator  (+, -, *, /) : ")

if op == "+":
    print(f"Result {a} {op} {b} : ", a + b)
elif op == "-":
    print("Result {a} {op} {b} : ", a - b)
elif op == "*":
    print(f"Result {a} {op} {b} : ", a * b)
elif op == "/":
    if b != 0:
        print(f"Result {a} {op} {b} : ", a / b)
    else:
        print(f"{a} {op} {b} : Cannot divide by zero")
else:
    print("Invalid operator")
```
